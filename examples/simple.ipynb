```python
import sys, os
sys.path.append(os.path.dirname(os.getcwd()))
from tsmnet import Stretcher
import torch
import torchaudio
import IPython.display as ipd
import soundfile as sf
    
stretcher = Stretcher('../scripts/logs/logs-musicnet/weights')
# try to change it to "logs-musicnet-512x"/"logs-musicnet-256x" and observe the difference
```

```python
x, sr = torchaudio.load(os.path.expanduser('~/Datasets/curated/moon.wav'))
x = torchaudio.transforms.Resample(orig_freq=sr, new_freq=22050)(x)
sr = 22050
```

```python
ipd.Audio(x, rate=sr)
```

```python
x_scaled = stretcher(x, 0.75)
```

```python
# torchaudio.save('slow.wav', torch.from_numpy(x_scaled), sr)
ipd.Audio(x_scaled, rate=sr)
```
